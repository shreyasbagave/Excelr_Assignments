```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python


df = pd.read_csv("sales_data_with_discounts.csv")


print(df.head())
```

```python

num_cols = df.select_dtypes(include=['float64', 'int64']).columns

```

```python

desc_stats = df[num_cols].describe().T 
print(desc_stats)
```

```python


median = df[num_cols].median()
mode = df[num_cols].mode().iloc[0]
std_dev = df[num_cols].std()
```

```python

print("\nMedian:\n", median)
print("\nMode:\n", mode)
print("\nStandard Deviation:\n", std_dev)
```

```python

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python

cat_cols = df.select_dtypes(include=['object']).columns


plt.figure(figsize=(8,5))
sns.countplot(x=df["City"])
plt.title("City-wise Distribution")
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(8,5))
sns.countplot(x=df["Brand"])
plt.title("Brand Distribution")
plt.xticks(rotation=45)
plt.show()
```

### Conclusion

Descriptive Analytics revealed that Avg Price has moderate variation, while Volume is highly dispersed.

Visualizations showed that Discount Amount has several outliers, possibly due to large-scale promotional campaigns.

Categorical analysis indicated that City X had the highest sales, while Brand Y dominated overall volume.
