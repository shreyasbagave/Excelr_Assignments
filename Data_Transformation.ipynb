```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder


```

```python
df=pd.read_csv('/content/Cardiotocographic.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

Some missing values in the dataset

```python


df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python


for col in df.select_dtypes(include="object"):
    df[col].fillna(df[col].mode()[0], inplace=True)
```

Numeric columns → imputed with mean (balances distribution).

Categorical columns → imputed with mode (most frequent).

```python


num_cols = df.select_dtypes(include=np.number).columns

```

```python


scaler_std = StandardScaler()
df_standard_scaled = pd.DataFrame(scaler_std.fit_transform(df[num_cols]), columns=num_cols)
```

```python

#  Min-Max Scaling
scaler_mm = MinMaxScaler()
df_minmax_scaled = pd.DataFrame(scaler_mm.fit_transform(df[num_cols]), columns=num_cols)
```

Standard Scaling → transforms data to mean=0, std=1.
 Best when data follows Gaussian distribution.
 Useful for algorithms like Logistic Regression, SVM, PCA.

```python

print("\nStandard Scaled Sample:\n", df_standard_scaled.head())
print("\nMin-Max Scaled Sample:\n", df_minmax_scaled.head())
```

```python



# One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=[col for col in df.columns if df[col].dtype == 'object' and df[col].nunique() < 5])
```

Min-Max Scaling → rescales values to [0,1].
 Preserves relative relationships.
Useful for algorithms like KNN, Neural Networks where distance matters.

```python

# Label Encoding
le = LabelEncoder()
for col in df.select_dtypes(include="object"):
    if df[col].nunique() > 5:
        df_encoded[col] = le.fit_transform(df[col])
```

One-Hot Encoding → used for categorical features with <5 categories.
 No ordinal assumption.

 Increases dimensionality if too many categories.


Label Encoding → used for categorical features with >5 categories.


Introduces ordinal meaning, which may mislead models if categories are unordered.

```python

print("\nEncoded Data Sample:\n", df_encoded.head())
```

```python
df['LB_AC_Ratio'] = df['LB'] / (df['AC'] + 1)   # +1 avoids division by zero
```

```python
df['Var_Index'] = df['ASTV'] * df['MSTV']
```

```python
df['UC_Log'] = np.log1p(df['UC'])

print("\nFeature Engineered Data Sample:\n", df[['LB', 'AC', 'LB_AC_Ratio', 'ASTV', 'MSTV', 'Var_Index', 'UC', 'UC_Log']].head())
```

LB_AC_Ratio

Captures relationship between baseline heart rate (LB) and accelerations (AC).

High or low ratios might indicate abnormal fetal health conditions.

 Var_Index

Combines ASTV (Avg Short Term Variability) and MSTV (Mean ST Variability).

Represents overall variability strength, important for detecting fetal distress.

UC_Log

Skewed data (like Uterine Contractions) is transformed with log.

Reduces skewness, helps algorithms like Linear Models learn better.
